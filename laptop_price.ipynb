```python
x=df1
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
lnr_model=LinearRegression()
knr_model=KNeighborsRegressor()
dtr_model=DecisionTreeRegressor()
rfr_model=RandomForestRegressor()
models=[lnr_model,knr_model,dtr_model,rfr_model]

```

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error,mean_absolute_error
accuracy_df=pd.DataFrame(columns=['model','score','mae','per'])
for i in models:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  score=r2_score(y_test,y_pred)
  mae=mean_absolute_error(y_test,y_pred)
  per=mean_absolute_percentage_error(y_test,y_pred)
  accuracy_df=accuracy_df.append({'model':str(i),'score':score,'mae':mae,'per':per},ignore_index=True)
accuracy_df
```

```python
accuracy_df.sort_values(by="score",ascending=False)
```

```python
plt.figure(figsize=(5,5))
plt.xticks(rotation=45)
ax=sns.barplot(x="model",y="score",palette="muted",
               data=accuracy_df.sort_values(by="score",ascending=False),errwidth=0)
for i in ax.containers:
  ax.bar_label(i,)
```
